```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.image as mpimg
from torch.utils.data.dataset import Dataset
from torch.utils.data.sampler import SubsetRandomSampler
from torch.utils.data import DataLoader
import matplotlib.patches as patches
import os
import math
from os import listdir
from os.path import isfile, join
from PIL import Image
from torch.autograd import Variable
import cv2
import torchvision.models as models

def truncated_normal_(tensor, mean=0, std=1):
    size = tensor.shape
    tmp = tensor.new_empty(size + (4,)).normal_()
    valid = (tmp < 2) & (tmp > -2)
    ind = valid.max(-1, keepdim=True)[1]
    tensor.data.copy_(tmp.gather(-1, ind).squeeze(-1))
    tensor.data.mul_(std).add_(mean)
    
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
from dogloader import dogs
input_size = 224
root_dir = "DATASETS"

input_transforms = transforms.Compose([
            transforms.Resize((input_size, input_size)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406],std=[0.229, 0.224, 0.225])])
train_dataset = dogs(root=root_dir,
                                 cropped = True,
                                 train=True,
                                 transform=input_transforms)
train_loader = DataLoader(train_dataset, batch_size=100, shuffle=True)
test_dataset = dogs(root=root_dir,
                                 cropped = True,
                                 train=False,
                                 transform=input_transforms)
test_loader = DataLoader(train_dataset, batch_size=100, shuffle=True)
```

```python
dataiter = iter(train_loader)
images, labels = dataiter.next()
plt.imshow(np.transpose(images[0].numpy(), (1,2,0)))
```

```python
# net = Net()
resnet50 = models.resnet50(pretrained=True)
net = resnet50
for param in net.parameters():
    param.requires_grad = False
num_ftrs = net.fc.in_features
net.fc = torch.nn.Linear(num_ftrs, 120)
net.to(device)
# print(net)


criterion = nn.CrossEntropyLoss()
# optimizer = optim.SGD(net.parameters(), lr=0.1, momentum=0.9)
optimizer = optim.Adam(net.parameters(), lr = 3e-3)
accuracy_array = []

for epoch in range(15): 

    running_loss = 0.0
    accuracy = 0
    accuracy_epoc = 0
    for i, data in enumerate(train_loader, 0):
        # get the inputs
        inputs, labels = data
        inputs, labels = inputs.to(device), labels.to(device)

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = net(inputs)
        maxs, indices = torch.max(outputs, 1)
        accuracy = sum(torch.eq(indices, labels)).float()/len(labels)
        
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        accuracy_epoc += accuracy
    avg_acc= accuracy_epoc/(i+1)
    accuracy_array.append(avg_acc)
    print('[%d, %5d] loss: %.3f accuracy: %.3f' %(epoch + 1, i + 1, running_loss / 500, avg_acc))
    running_loss = 0.0

    if avg_acc >= 0.99:
        break
print('Finished Training')
plt.plot(accuracy_array)
```

## AE

```python
from dogloader import dogs

input_size = 128
root_dir = "DATASETS"

input_transforms = transforms.Compose([
            transforms.Resize((input_size, input_size)),
            transforms.ToTensor()])
train_dataset = dogs(root=root_dir,
                                 cropped = True,
                                 train=True,
                                 transform=input_transforms)
train_loader = DataLoader(train_dataset, batch_size=100, shuffle=True)
test_dataset = dogs(root=root_dir,
                                 cropped = True,
                                 train=False,
                                 transform=input_transforms)
test_loader = DataLoader(train_dataset, batch_size=100, shuffle=True)
```

```python

class AutoEncoder(nn.Module):
    def __init__(self):
        super(AutoEncoder, self).__init__()

        self.encode_conv1 = nn.Conv2d(3, 8, 3, stride=2, padding = 1)
        self.conv1_bn = nn.BatchNorm2d(8)
        self.encode_conv2 = nn.Conv2d(8, 16, 3, stride=2, padding = 1)
        self.conv2_bn = nn.BatchNorm2d(16)
        self.encode_conv3 = nn.Conv2d(16, 32, 3, stride=2, padding = 1)
        self.conv3_bn = nn.BatchNorm2d(32)
        self.encode_conv4 = nn.Conv2d(32, 64, 3, stride=2, padding = 1)
        self.conv4_bn = nn.BatchNorm2d(64)
        
        self.ln_encode = nn.Linear(4096, 256)
        
        self.decode_conv1 = nn.Conv2d(1, 64, 3, stride=2, padding = 1)
        self.conv1_decode_bn = nn.BatchNorm2d(64)
        self.decode_trans_conv1 = nn.ConvTranspose2d(64, 32, 3, stride=3, padding=1)
        self.conv1_trans_bn = nn.BatchNorm2d(32)
        self.decode_trans_conv2 = nn.ConvTranspose2d(32, 16, 3, stride=3, padding=1)
        self.conv2_trans_bn = nn.BatchNorm2d(16)
        self.decode_trans_conv3 = nn.ConvTranspose2d(16, 8, 4, stride=2, padding=1)
        self.conv3_trans_bn = nn.BatchNorm2d(8)
        self.decode_conv2 = nn.Conv2d(8, 3, 3, stride=1, padding = 1)
        

    def encode(self, x):
        x = F.relu(self.conv1_bn(self.encode_conv1(x)))
        x = F.relu(self.conv2_bn(self.encode_conv2(x)))
        x = F.relu(self.conv3_bn(self.encode_conv3(x)))
        x = F.relu(self.conv4_bn(self.encode_conv4(x)))
        return self.ln_encode(x.view(-1,4096))

    def decode(self, z):
        z = F.relu(self.conv1_decode_bn(self.decode_conv1(z)))
        z = F.relu(self.conv1_trans_bn(self.decode_trans_conv1(z)))
        z = F.relu(self.conv2_trans_bn(self.decode_trans_conv2(z)))
        z = F.relu(self.conv3_trans_bn(self.decode_trans_conv3(z)))
        z = self.decode_conv2(z)
        return torch.sigmoid(z)

    def forward(self, x):
        code = self.encode(x)
        return self.decode(code.view(-1,1,16,16))


model = AutoEncoder().to(device)
optimizer = optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.MSELoss()
train_array = []
```

```python
for i in range(100):
    train_loss = 0
    for batch_idx, (data, _) in enumerate(train_loader):
        data = data.to(device)
        optimizer.zero_grad()
        recon = model(data)
        loss = criterion(recon, data)
        loss.backward()
        train_loss += loss.item()
        optimizer.step()
        
    train_array.append(train_loss)
    if i % 5 == 0:
        print("Train Epoch: {} Loss: {:.6f}".format(i,train_loss/batch_idx))
        img = np.transpose(recon[0].cpu().detach().numpy(), (1,2,0))
        plt.imshow(img)
        plt.show()
```

```python
class AENet(nn.Module):
    def __init__(self, ae):
        super(AENet, self).__init__()
        self.ae = ae
        self.fc1 = nn.Linear(256,120)
    def forward(self, x):
        code = self.ae.encode(x)
        x = self.fc1(x)
        return F.softmax(x)
```

```python
for param in ae.parameters():
    param.requires_grad = False
aenet = AENet(ae)
aenet.to(device)


criterion = nn.CrossEntropyLoss()
# optimizer = optim.SGD(net.parameters(), lr=0.1, momentum=0.9)
optimizer = optim.Adam(aenet.parameters(), lr = 3e-3)
accuracy_array = []

for epoch in range(15): 

    running_loss = 0.0
    accuracy = 0
    accuracy_epoc = 0
    for i, data in enumerate(train_loader, 0):
        # get the inputs
        inputs, labels = data
        inputs, labels = inputs.to(device), labels.to(device)

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = aenet(inputs)
        maxs, indices = torch.max(outputs, 1)
        accuracy = sum(torch.eq(indices, labels)).float()/len(labels)
        
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        accuracy_epoc += accuracy
    avg_acc= accuracy_epoc/(i+1)
    accuracy_array.append(avg_acc)
    print('[%d, %5d] loss: %.3f accuracy: %.3f' %(epoch + 1, i + 1, running_loss / 500, avg_acc))
    running_loss = 0.0

    if avg_acc >= 0.99:
        break
print('Finished Training')
plt.plot(accuracy_array)
```

## VAE

```python
from dogloader import dogs

input_size = 128
root_dir = "DATASETS"

input_transforms = transforms.Compose([
            transforms.Resize((input_size, input_size)),
            transforms.ToTensor()])
train_dataset = dogs(root=root_dir,
                                 cropped = True,
                                 train=True,
                                 transform=input_transforms)
train_loader = DataLoader(train_dataset, batch_size=100, shuffle=True)
test_dataset = dogs(root=root_dir,
                                 cropped = True,
                                 train=False,
                                 transform=input_transforms)
test_loader = DataLoader(train_dataset, batch_size=100, shuffle=True)
```

```python
epochs = 10
interval = 1
    
class VAE(nn.Module):
    def __init__(self):
        super(VAE, self).__init__()

        self.encode_conv1 = nn.Conv2d(3, 8, 3, stride=2, padding = 1)
        self.conv1_bn = nn.BatchNorm2d(8)
        self.encode_conv2 = nn.Conv2d(8, 16, 3, stride=2, padding = 1)
        self.conv2_bn = nn.BatchNorm2d(16)
        self.encode_conv3 = nn.Conv2d(16, 32, 3, stride=2, padding = 1)
        self.conv3_bn = nn.BatchNorm2d(32)
        self.encode_conv4 = nn.Conv2d(32, 64, 3, stride=2, padding = 1)
        self.conv4_bn = nn.BatchNorm2d(64)
        
        self.ln_encode_mean = nn.Linear(4096, 256)
        self.ln_encode_variance = nn.Linear(4096, 256)
        self.ln_decode_variance = nn.Linear(256, 4096)
        
        self.decode_trans_conv1 = nn.ConvTranspose2d(4096, 512, 5, stride=2)
        self.conv1_trans_bn = nn.BatchNorm2d(512)
        self.decode_trans_conv2 = nn.ConvTranspose2d(512, 256, 5, stride=5, padding = 1)
        self.conv2_trans_bn = nn.BatchNorm2d(256)
        self.decode_trans_conv3 = nn.ConvTranspose2d(256, 128, 3, stride=3, padding = 2)
        self.conv3_trans_bn = nn.BatchNorm2d(128)
        self.decode_trans_conv4 = nn.ConvTranspose2d(128, 3, 2, stride=2, padding = 1)
        self.sigmoid = nn.Sigmoid()

    def encode(self, x):
        x = F.relu(self.conv1_bn(self.encode_conv1(x)))
        x = F.relu(self.conv2_bn(self.encode_conv2(x)))
        x = F.relu(self.conv3_bn(self.encode_conv3(x)))
        x = F.relu(self.conv4_bn(self.encode_conv4(x)))
        
        mean, log_variance = self.ln_encode_mean(x.view(-1,4096)), self.ln_encode_variance(x.view(-1,4096))
        std = torch.exp(0.5*log_variance)
        ns = torch.randn_like(std) #return N(0,1)
        z = ns * std + mean
        
        return z, mean, log_variance

    def decode(self, z):
        z = self.ln_decode_variance(z)
        z = F.relu(self.conv1_trans_bn(self.decode_trans_conv1(z.view(-1,4096,1,1))))
#         print(z.shape)
        z = F.relu(self.conv2_trans_bn(self.decode_trans_conv2(z)))
#         print(z.shape)
        z = F.relu(self.conv3_trans_bn(self.decode_trans_conv3(z)))
#         print(z.shape)
        z = self.decode_trans_conv4(z)
#         print(z.shape)
        return self.sigmoid(z)

    def forward(self, x):
        z, mean, log_variance = self.encode(x)
        return self.decode(z), mean, log_variance


    
del vae
vae = VAE().to(device)
optimizer = optim.Adam(vae.parameters(), lr=1e-3)
```

```python
def loss_function(recon_x, x, mean, logvariance):
    BCE = F.binary_cross_entropy(recon_x, x.view(-1, 49152), reduction='sum')
    KLD = 0.5 * torch.sum(mean.pow(2) + logvariance.exp() - logvariance - 1)

    return BCE + KLD


def train(epochs):
    loss_array = []
    for epoch in range(epochs):
        train_loss = 0
        for batch_idx, (data, _) in enumerate(train_loader):
            data = data.to(device)
            
            optimizer.zero_grad()
            recon, mu, logvariance = vae(data)
            loss = loss_function(recon, data, mu, logvariance)
            loss.backward()
            train_loss += loss.item()
            optimizer.step()
            if data.shape[0]>=5 :
                vis = recon
        if epoch % interval == 0:
            f, axarr = plt.subplots(1, 5)
            f.set_figheight(5)
            f.set_figwidth(10)

            for i in range(5):
                imgorg = np.transpose(vis[i].cpu().detach().numpy(), (1,2,0))
                axarr[i].imshow(imgorg)
            plt.show()  
            
            print('====> Epoch: {} Average loss: {:.4f}'.format(
                  epoch, train_loss / len(train_loader.dataset)))
        
        loss_array.append(train_loss / len(train_loader.dataset))
    return loss_array


loss_array = train(epochs)
```
